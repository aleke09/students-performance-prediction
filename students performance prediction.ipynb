# selecting important featuers for training

```python
df = student_mat[[ 'G1', 'G2', 'G3', 'failures', 'studytime', 'absences', 'age']]
y = df[['G3']]
```

```python
df.sample(50)
```

# dividing the dataset into input variable and target 

```python
X = np.array(df.drop(['G3'], axis=1))
y = np.array(df['G3'])
```

# Training the model using Decission tree clasifier()

```python
# clf = DecisionTreeClassifier()
# clf = sklearn.svm.SVC(kernel='linear')
clf = sklearn.neighbors.KNeighborsClassifier()
x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(X, y, test_size=.2)
clf.fit(x_train, y_train)
predictions = clf.predict(x_test)
acc = clf.score(x_test, y_test)
score = accuracy_score(y_test, predictions)
print(acc)
print(predictions)
print(score)
```
